```
# Dictionaries
# What's a dictionary? {Key: value} pair is called Dictionary
```

```
# int = 4
# float = 6.8
# str = "Hasham"
# bool = True / False
# lists = [1, 2, 3, 4, 5, 6]
# dict
```

```
# dict_syntax = { 
#     "key" : "value",
#     "key2" : "value2",
# }
```

```
person = {
    "name" : "Hasham", 
    "age"  : 18,
    "student" : True,
    "height" : 6.0
}
```

```
person
```

```
for i in person: #keys --> #name, age, student ,height
  print(i)
```

```
person['age'] # dict_name['key'] --> value
```

```
person['height'] #value 6.0
```

```
for i in person: # keys
  print(f'key is : {i} ,whose value is: {person[i]}')
```

```
person['age'] = 19
```
